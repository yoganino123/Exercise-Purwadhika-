# OCR Libraries Comparison: EasyOCR, Tesseract, and DocTR

## Overview

This notebook provides comprehensive examples and comparisons of three popular OCR libraries:

- **Developer**: JaidedAI

- **Developer**: Google (engine) + Python wrapper

- **Best for**: Document OCR, clean text images

### 3. [DocTR](https://github.com/mindee/doctr)
- **Developer**: Mindee
- **Architecture**: Modern deep learning pipeline
- **Focus**: Document understanding and layout analysis
- **Best for**: Document processing, production applications

````python
# Import required libraries for EasyOCR
import easyocr
import numpy as np
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt

print("EasyOCR Examples - GitHub: https://github.com/JaidedAI/EasyOCR")
print("Installation: pip install easyocr")
print("Note: First run downloads models (~100-500MB per language)")
````

````python
def basic_easyocr_demo():
    """Basic EasyOCR usage example.

    Returns:
        list: OCR results with bounding boxes, text, and confidence scores.
    """

    print("EasyOCR Basic Usage:")
    print("=" * 30)

    # Initialize reader (downloads models on first run)
    reader = easyocr.Reader(['en'])  # English only

    # Read text from image
    # Accepts: file path, URL, numpy array, PIL Image, or bytes
    results = reader.readtext('image.jpg')

    # Process results
    # Format: [(bbox, text, confidence), ...]
    for (bbox, text, confidence) in results:
        print(f"Text: '{text}'")
        print(f"Confidence: {confidence:.3f}")
        print(f"Bounding Box: {bbox}")
        print("-" * 20)

    return results
````

````python
def advanced_easyocr_params():
    """Advanced EasyOCR parameters for fine-tuning.

    Returns:
        list: OCR results with advanced parameter configuration.
    """

    print("\nAdvanced Parameters:")
    print("=" * 30)

    reader = easyocr.Reader(['en'])

    # Advanced parameter example
    results = reader.readtext(
        'image.jpg',
        # Detection parameters
        width_ths=0.7,          # Text box width threshold
        height_ths=0.7,         # Text box height threshold

        # Recognition parameters
        decoder='greedy',       # 'greedy' or 'beamsearch'
        beamWidth=5,           # Beam search width

        # Filtering
        paragraph=False,        # Group into paragraphs
        min_size=20,           # Minimum text area
        text_threshold=0.7,    # Text confidence threshold
        low_text=0.4,          # Low text score
        link_threshold=0.4,    # Link threshold
        canvas_size=2560,      # Max image size
        mag_ratio=1.0,         # Magnification ratio

        # Output format
        detail=1               # 0: text only, 1: full details
    )

    print(f"Parameters configured for enhanced accuracy")
    return results
````

````python
# Import required libraries for Tesseract
import pytesseract
from PIL import Image, ImageDraw, ImageFont
import cv2
import numpy as np

print("Tesseract Examples - GitHub: https://github.com/madmaze/pytesseract")
print("Installation: pip install pytesseract")
print("Requires: Tesseract engine installed on system")
````

````python
def basic_tesseract_demo():
    """Basic Tesseract usage examples.

    Demonstrates simple text extraction and word-level analysis
    with confidence scores and bounding boxes.
    """

    print("Tesseract Basic Usage:")
    print("=" * 30)

    # Load image
    image = Image.open('sample.jpg')

    # Simple text extraction
    text = pytesseract.image_to_string(image)
    print(f"Extracted text: {text}")

    # Text with confidence scores
    data = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

    # Process word-level results
    for i, word in enumerate(data['text']):
        if word.strip():  # Skip empty strings
            confidence = data['conf'][i]
            x, y, w, h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
            print(f"Word: '{word}' | Confidence: {confidence} | BBox: ({x},{y},{w},{h})")
````

````python
def tesseract_languages():
    """Tesseract language support.

    Demonstrates single and multi-language text recognition
    with different language configurations.
    """

    print("\nLanguage Support:")
    print("=" * 30)

    image = Image.open('sample.jpg')

    # Single language
    text_english = pytesseract.image_to_string(image, lang='eng')
    text_german = pytesseract.image_to_string(image, lang='deu')
    text_chinese = pytesseract.image_to_string(image, lang='chi_sim')
    text_japanese = pytesseract.image_to_string(image, lang='jpn')

    # Multiple languages
    text_multi = pytesseract.image_to_string(image, lang='eng+deu+fra')

    # Common language codes
    languages = {
        'eng': 'English',
        'deu': 'German',
        'fra': 'French',
        'spa': 'Spanish',
        'chi_sim': 'Chinese Simplified',
        'chi_tra': 'Chinese Traditional',
        'jpn': 'Japanese',
        'kor': 'Korean',
        'ara': 'Arabic',
        'hin': 'Hindi',
        'rus': 'Russian'
    }

    print("Common Language Codes:")
    for code, name in languages.items():
        print(f"  {code}: {name}")
````

````python
def tesseract_configuration():
    """Advanced Tesseract configuration options.

    Returns:
        str: Extracted text using custom configuration.
    """

    print("\nOCR Configuration:")
    print("=" * 30)

    image = Image.open('sample.jpg')

    # Page Segmentation Modes (PSM)
    psm_modes = {
        0: 'Orientation and script detection (OSD) only',
        1: 'Automatic page segmentation with OSD',
        2: 'Automatic page segmentation, but no OSD, or OCR',
        3: 'Fully automatic page segmentation, but no OSD (Default)',
        4: 'Assume a single column of text of variable sizes',
        5: 'Assume a single uniform block of vertically aligned text',
        6: 'Assume a single uniform block of text',
        7: 'Treat the image as a single text line',
        8: 'Treat the image as a single word',
        9: 'Treat the image as a single word in a circle',
        10: 'Treat the image as a single character',
        11: 'Sparse text. Find as much text as possible in no particular order',
        12: 'Sparse text with OSD',
        13: 'Raw line. Treat the image as a single text line'
    }

    # OCR Engine Modes (OEM)
    oem_modes = {
        0: 'Legacy engine only',
        1: 'Neural nets LSTM engine only',
        2: 'Legacy + LSTM engines',
        3: 'Default, based on what is available'
    }

    # Configuration examples
    configs = [
        '--oem 3 --psm 6',              # Default
        '--oem 1 --psm 7',              # LSTM + single line
        '--oem 3 --psm 8',              # Single word
        '--oem 3 --psm 6 -c preserve_interword_spaces=1',  # Preserve spaces
        '--oem 3 --psm 6 -c tessedit_char_whitelist=0123456789'  # Numbers only
    ]

    # Example usage
    custom_config = r'--oem 3 --psm 6'
    text = pytesseract.image_to_string(image, config=custom_config)

    print("Configuration applied successfully")
    return text
````

````python
def tesseract_bounding_boxes():
    """Extract text with bounding boxes.

    Returns:
        list: Word-level data with positions and confidence scores.
    """

    print("\nBounding Box Extraction:")
    print("=" * 30)

    image = Image.open('sample.jpg')

    # Character-level bounding boxes
    boxes = pytesseract.image_to_boxes(image)
    print("Character boxes format: char x1 y1 x2 y2 page")
    print(boxes[:100])  # Show first 100 characters

    # Word-level data with positions
    data = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

    # Extract word positions
    word_data = []
    for i, word in enumerate(data['text']):
        if word.strip() and int(data['conf'][i]) > 30:  # Filter low confidence
            word_info = {
                'text': word,
                'x': data['left'][i],
                'y': data['top'][i],
                'width': data['width'][i],
                'height': data['height'][i],
                'confidence': data['conf'][i]
            }
            word_data.append(word_info)

    return word_data
````

````python
def tesseract_preprocessing():
    """Image preprocessing for better OCR results.

    Demonstrates various image preprocessing techniques
    to improve OCR accuracy.
    """

    print("\nImage Preprocessing:")
    print("=" * 30)

    # Load image with OpenCV
    image = cv2.imread('sample.jpg')

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Apply different preprocessing techniques

    # 1. Thresholding
    _, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    # 2. Noise removal
    denoised = cv2.medianBlur(gray, 5)

    # 3. Morphological operations
    kernel = np.ones((1,1), np.uint8)
    opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)

    # 4. Dilation and erosion
    dilated = cv2.dilate(opening, kernel, iterations=1)

    # Run OCR on preprocessed images
    text_original = pytesseract.image_to_string(gray)
    text_thresh = pytesseract.image_to_string(thresh)
    text_denoised = pytesseract.image_to_string(denoised)

    print("Preprocessing techniques applied:")
    print("- Grayscale conversion")
    print("- OTSU thresholding")
    print("- Noise removal")
    print("- Morphological operations")
````

````python
# Import required libraries for DocTR
from doctr.models import ocr_predictor
from doctr.io import DocumentFile
import matplotlib.pyplot as plt
import numpy as np

print("DocTR Examples - GitHub: https://github.com/mindee/doctr")
print("Installation: pip install python-doctr[torch]")
print("Features: Document OCR, Layout analysis, Production-ready")
````

````python
def basic_doctr_demo():
    """Basic DocTR usage for document OCR.

    Returns:
        DocTR result object with OCR results.
    """

    print("DocTR Basic Usage:")
    print("=" * 30)

    # Load the OCR predictor
    model = ocr_predictor(pretrained=True)

    # Load document (supports images and PDFs)
    doc = DocumentFile.from_images(['sample.jpg'])

    # Perform OCR
    result = model(doc)

    # Extract text
    extracted_text = result.render()
    print(f"Extracted text: {extracted_text}")

    # Get structured output
    json_output = result.export()
    print(f"Pages processed: {len(json_output['pages'])}")

    return result
````

````python
def doctr_document_analysis():
    """Analyze document structure with DocTR.

    Performs hierarchical analysis of document structure including
    pages, blocks, lines, and words.
    """

    print("\nDocument Structure Analysis:")
    print("=" * 35)

    model = ocr_predictor(pretrained=True)
    doc = DocumentFile.from_images(['document.jpg'])
    result = model(doc)

    # Analyze document hierarchy
    for page_idx, page in enumerate(result.pages):
        print(f"\nPage {page_idx + 1}:")
        print(f"  Blocks: {len(page.blocks)}")

        for block_idx, block in enumerate(page.blocks):
            print(f"  Block {block_idx + 1}:")
            print(f"    Lines: {len(block.lines)}")

            for line_idx, line in enumerate(block.lines):
                print(f"    Line {line_idx + 1}:")
                print(f"      Words: {len(line.words)}")

                # Extract word details
                line_text = ""
                for word in line.words:
                    line_text += word.value + " "
                    print(f"        Word: '{word.value}' | Confidence: {word.confidence:.3f}")

                print(f"      Line text: {line_text.strip()}")
````

````python
def doctr_pdf_processing():
    """Process multi-page PDF documents.

    Returns:
        list: Text content from each page of the PDF.
    """

    print("\nPDF Processing:")
    print("=" * 20)

    model = ocr_predictor(pretrained=True)

    # Load PDF document
    doc = DocumentFile.from_pdf('document.pdf')
    result = model(doc)

    # Process each page
    all_text = []
    for page_idx, page in enumerate(result.pages):
        page_text = ""

        for block in page.blocks:
            for line in block.lines:
                for word in line.words:
                    page_text += word.value + " "
                page_text += "\n"  # New line after each text line

        all_text.append(page_text)
        print(f"Page {page_idx + 1} processed: {len(page_text)} characters")

    return all_text
````

````python
def doctr_custom_models():
    """Using different detection and recognition models.

    Demonstrates how to configure custom detection and recognition
    architectures for specific use cases.
    """

    print("\nCustom Models:")
    print("=" * 20)

    # Default predictor
    default_model = ocr_predictor(pretrained=True)

    # Custom detection and recognition models
    custom_model = ocr_predictor(
        det_arch='db_resnet50',           # Detection architecture
        reco_arch='crnn_vgg16_bn',        # Recognition architecture
        pretrained=True
    )

    # Available architectures
    detection_models = [
        'db_resnet50',
        'db_mobilenet_v3_large',
        'linknet_resnet18',
        'linknet_resnet34'
    ]

    recognition_models = [
        'crnn_vgg16_bn',
        'crnn_mobilenet_v3_small',
        'sar_resnet31',
        'master'
    ]

    print("Available Detection Models:")
    for model in detection_models:
        print(f"  - {model}")

    print("\nAvailable Recognition Models:")
    for model in recognition_models:
        print(f"  - {model}")
````

````python
def doctr_batch_processing():
    """Efficient batch processing of multiple documents.

    Returns:
        list: Batch results with text and metadata for each document.
    """

    print("\nBatch Processing:")
    print("=" * 20)

    model = ocr_predictor(pretrained=True)

    # Process multiple images
    image_paths = ['doc1.jpg', 'doc2.jpg', 'doc3.jpg']
    doc = DocumentFile.from_images(image_paths)

    # Batch processing
    result = model(doc)

    # Extract results for each document
    batch_results = []
    for page_idx, page in enumerate(result.pages):
        page_text = ""
        word_count = 0

        for block in page.blocks:
            for line in block.lines:
                for word in line.words:
                    page_text += word.value + " "
                    word_count += 1

        batch_results.append({
            'image': image_paths[page_idx] if page_idx < len(image_paths) else f'page_{page_idx}',
            'text': page_text.strip(),
            'word_count': word_count
        })

    return batch_results
````

````python
def doctr_quality_analysis():
    """Analyze OCR quality and confidence scores.

    Returns:
        tuple: (confidence_scores, low_confidence_words) for quality analysis.
    """

    print("\nQuality Analysis:")
    print("=" * 20)

    model = ocr_predictor(pretrained=True)
    doc = DocumentFile.from_images(['sample.jpg'])
    result = model(doc)

    # Analyze confidence scores
    confidence_scores = []
    low_confidence_words = []

    for page in result.pages:
        for block in page.blocks:
            for line in block.lines:
                for word in line.words:
                    confidence = word.confidence
                    confidence_scores.append(confidence)

                    if confidence < 0.7:  # Low confidence threshold
                        low_confidence_words.append({
                            'word': word.value,
                            'confidence': confidence,
                            'geometry': word.geometry
                        })

    # Calculate statistics
    if confidence_scores:
        avg_confidence = np.mean(confidence_scores)
        min_confidence = np.min(confidence_scores)
        max_confidence = np.max(confidence_scores)

        print(f"Average confidence: {avg_confidence:.3f}")
        print(f"Min confidence: {min_confidence:.3f}")
        print(f"Max confidence: {max_confidence:.3f}")
        print(f"Low confidence words: {len(low_confidence_words)}")

    return confidence_scores, low_confidence_words
````
