## Use GPT Model for Text Generation

```python
%%time
model_id = "gpt-4o-mini"

response = openai.chat.completions.create(
    model=model_id,
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "Explain how gravity works in simple terms."}
    ]
)

print(response.choices[0].message.content)
```

## Use Embeddings for Text Similarity
You can compare embeddings using cosine similarity:

### Generate Embeddings (Text to Vector)

```python
model_id = "text-embedding-3-small"

embedding_response = openai.embeddings.create(
    input=["Indonesia has many islands.", "Soccer is a popular sport in Europe."],
    model=model_id)

embeddings = [item.embedding for item in embedding_response.data]

print(f"Vector length: {len(embeddings[0])}")
```

```python
embedding_response = openai.embeddings.create(
    input=["Indonesia has many islands.", "Soccer is a popular sport in Europe.", "Soccer also popular in Indonesia"],
    model=model_id)

embeddings = [item.embedding for item in embedding_response.data]
len(embeddings)
```

### Compare the Vectors

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import openai

# Example inputs
texts = ["I love playing football.", "Soccer is my favorite sport.", "Cooking is relaxing."]

# Get embeddings
embedding_response = openai.embeddings.create(input=texts, model="text-embedding-3-small")
embeddings = [item.embedding for item in embedding_response.data]

# Calculate similarity between first and second texts
similarity = cosine_similarity([embeddings[0]], [embeddings[1]])
print(f"Similarity between 1 and 2: {similarity[0][0]:.4f}")
```

```python
# Calculate similarity between first and second texts
similarity = cosine_similarity([embeddings[0]], [embeddings[2]])
print(f"Similarity between 1 and 3: {similarity[0][0]:.4f}")
```

```python
# Calculate similarity between first and second texts
similarity = cosine_similarity([embeddings[1]], [embeddings[2]])
print(f"Similarity between 2 and 3: {similarity[0][0]:.4f}")
```
