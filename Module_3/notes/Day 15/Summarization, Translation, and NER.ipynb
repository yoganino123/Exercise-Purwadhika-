## Extractive Summarization Prompts

```python
sys_prompt = """You are an expert extractive summarizer. Your task is to create a summary by selecting and combining the most important sentences from the original text WITHOUT paraphrasing or rewriting them.

Instructions:
1. Identify the 3-5 most important sentences that capture the key information
2. Extract these sentences EXACTLY as they appear in the original text
3. Arrange them in logical order to form a coherent summary
4. Do not add, modify, or rephrase any words

Text to summarize:
{input_text}

Output format:
SELECTED SENTENCES:
- [Sentence 1 from original text]
- [Sentence 2 from original text]
- [Sentence 3 from original text]

EXTRACTIVE SUMMARY:
[Combine the selected sentences into a flowing summary]
"""
```

```python
messages = [
    SystemMessage(sys_prompt),
    HumanMessage(llm_article),
]
response = model.invoke(messages)
display(Markdown(response.content))
```

### Sales Call Summary

```python
CALL_DATA = {
  "id": 102,
  "agent_id": 100,
  "customer_id": 1001,
  "call_history": [
    {"role": "agent", "message": "Hello, this is Daniel from GreenEnergy Solutions. How are you today?"},
    {"role": "user", "message": "I’m busy. What is this about?"},
    {"role": "agent", "message": "I’ll be quick. We’re helping businesses reduce energy costs by switching to our solar packages. Are you currently happy with your energy bills?"},
    {"role": "user", "message": "Honestly, my bills are high but I don’t have time for this sales pitch."},
    {"role": "agent", "message": "I completely understand. Just to give you an idea, many of our clients cut costs by 20%. Could I schedule a quick call to show you the numbers?"},
    {"role": "user", "message": "Did you not hear me? I said I don’t have time."},
    {"role": "agent", "message": "It’ll only take five minutes, and it could save your company thousands of dollars."},
    {"role": "user", "message": "This is the third time your company has called me this month. It’s getting annoying."},
    {"role": "agent", "message": "I apologize, but I think you’ll find our offer valuable. We can install solar panels with no upfront cost—"},
    {"role": "user", "message": "Stop interrupting me! I told you I’m not interested."},
    {"role": "agent", "message": "If I could just finish—"},
    {"role": "user", "message": "No, you’re not listening! You keep calling during my workday. This is harassment."},
    {"role": "agent", "message": "Sir, I’m only trying to help reduce your expenses—"},
    {"role": "user", "message": "Help? You’re wasting my time. Every minute on this call costs me money."},
    {"role": "agent", "message": "I assure you this won’t take long. Many business owners thank us later."},
    {"role": "user", "message": "I don’t care what other business owners do. Take me off your call list."},
    {"role": "agent", "message": "Are you sure? This offer is limited and might not be available later."},
    {"role": "user", "message": "Enough! If you call me again, I will file a complaint with the consumer protection agency."},
    {"role": "agent", "message": "Understood. I’ll mark your number as do not contact."},
    {"role": "user", "message": "Good. Don’t ever call me again."},
    {"role": "agent", "message": "Apologies for the inconvenience. Goodbye."},
    {"role": "user", "message": "Finally."}
  ]
}
```

```python
def format_chat(chat_data):
  formatted = []
  for message in chat_data['call_history']:
    formatted.append(Document(page_content=message['message'],
                              metadata={'role': "customer" if message['role'] == "user" else "sales"}))
  return formatted

formatted_call_data = format_chat(CALL_DATA)
print(formatted_call_data)
```

```python
user_prompt = f"""Write a concise summary of the following:\n\n
{str(formatted_call_data)}

CONCISE SUMMARY:
"""
response = model.invoke(user_prompt)
display(Markdown(response.content))
```

### Structuring Summary with Structured Output

```python
from pydantic import BaseModel, Field
from langchain.agents import create_agent


class ContactInfo(BaseModel):
    """Contact information for a person."""
    title: str = Field(description="The title of article.")
    topic: str = Field(description="The topic of article.")
    summary: str = Field(description="Write a compact summary from article.")

agent = create_agent(
    model="gpt-4o-mini",
    response_format=ContactInfo  # Auto-selects ProviderStrategy
)

result = agent.invoke({
    "messages": [
        {"role": "system", "content": llm_article},
        {"role": "user", "content": "Extract information into given structure."},
                 ]
})
```

```python
title = result["structured_response"].title
topic = result["structured_response"].topic
summary = result["structured_response"].summary
```

```python
print(f"Title: {title}")
print(f"Topic: {topic}")
print(f"Summary: {summary}")
```
