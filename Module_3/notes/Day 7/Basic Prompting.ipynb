```python
import openai
from openai import OpenAI
from google.colab import userdata
from IPython.display import Markdown, display

import os

openai.api_key = userdata.get("OPENAI_API_KEY")
client = OpenAI(api_key=openai.api_key)
```

```python
model = "gpt-4o-mini"
```

```python
# jawab bahasa indonesia
messages = [
    {"role": "system", "content": "Kamu adalah AI Assistant yang ramah dan ceria. Kamu selalu menjawab pertanyaan dengan sopan dan ramah. Gunakan bahasa Indonesia dan gunakan emoji apabila diperlukan dalam percakapan."},
    {"role": "user", "content": "Saya ini orang awam, tolong dong jelasin tentang LLM."}
    ]

response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=0.7
)
```

```python
display(Markdown(response.choices[0].message.content))
```

```python
# jawab dengan bahasa jawa
messages = [
    {"role": "system", "content": "Kamu adalah AI Assistant yang ramah dan ceria. Kamu selalu menjawab pertanyaan dengan sopan dan ramah. Gunakan bahasa Jawa dan gunakan emoji apabila diperlukan dalam percakapan."},
    {"role": "user", "content": "Saya ini orang awam, tolong dong jelasin tentang LLM."}
    ]

response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=0.7
)
```

```python
# AI yang jutek
messages = [
    {"role": "system", "content": "Kamu adalah AI Assistant yang galak dan jutek. Kamu selalu marah-marah dalam menjawab pertanyaan dan menyampaikan jawaban. Gunakan bahasa non-formal dalam menjawab pertanyaan serta gunakan emoji apabila diperlukan."},
    {"role": "user", "content": "Saya ini orang awam, tolong dong jelasin tentang LLM woy!!"}
    ]

response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=0.7
)
```

```python
# zero-shot prompting
prompt = f'''Klasifikasikan kalimat berikut ke dalam kategori berikut:
- positif
- negatif
Text: Saya sangat senang hari ini.
Jawab langsung dengan label kategori yang tepat.
'''

messages = [
    {"role": "user", "content": prompt}
    ]

response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=0.7
)
```

```python
display(Markdown(response.choices[0].message.content))
```

```python
# few-shot prompting
prompt = f'''Klasifikasikan kalimat berikut ke dalam kategori berikut:
- positif
- negatif

text : Saya sangat senang hari ini.
lable : positif

text : Saya sangat sedih hari ini.
lable : negatif

text : Saya sangat marah hari ini.
lable :
'''

messages = [
    {"role": "user", "content": prompt}
    ]

response = client.chat.completions.create(
    model=model,
    messages=messages,
    temperature=0.7
)
```

```python
display(Markdown(response.choices[0].message.content))
```
