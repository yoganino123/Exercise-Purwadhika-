```python
# importing all libraries needed
import pandas as pd #for data processing
from langchain_text_splitters import CharacterTextSplitter, RecursiveCharacterTextSplitter #for splitter (chunking) method
from langchain_core.documents import Document #for langchain Document object
```

#Load dataset

```python
#directory to dataset
data_path = "/content/drive/MyDrive/JCAIEAH-003/Notes and Hands On/Modul 3/Day 11/bbc-news-data.csv"
```

```python
#read dataset with pandas
df = pd.read_csv(data_path, sep='\t')
print(df.shape) #show total row and column with format (row, column)
df.head() #show first 5 record of dataset
```

```python
df['category'].value_counts() #show category amount
```

```python
#check characters amount is first news content
print(f"Amount of characters: {len(df.content[0])}")
#show the content
print(df.content[0])
```

#Chunking

```python
#fixed size chunking
character_size = 101 #each chunk contains 101 characters
char_count=0 #character counter for while loop incremental
selected_content = df.content[0] #select a news content from dataset
while char_count < len(df.content[0]):
  print(selected_content[char_count:char_count+character_size])
  print("=======================================================================================================================")
  char_count += character_size
```

```python
# store to langchain Document object
character_size = 101 #each chunk contains 101 characters
char_count=0 #character counter for while loop incremental
selected_content = df.content[0] #select a news content from dataset
while char_count < len(df.content[0]):
  print(Document(page_content=selected_content[char_count:char_count+character_size]))
  char_count += character_size
```

```python
#chunk text with CharacterTextSplitter
#separator: " " (space)
#chunk size: 101 characters
#chunk overlap: 10 (current chunk will bring last 10 characters from previous chunk)
splitter = CharacterTextSplitter(
    separator=" ",
    chunk_size=101,
    chunk_overlap=10,
)
```

```python
splitter.split_text(selected_content)
```

```python
#split the text and transform into langchain Document object
splitter.create_documents([selected_content])
```

```python
#chunk text with CharacterTextSplitter
#separators: [" ", "."] (space, dot)
#chunk size: 101 characters
#chunk overlap: 10 (current chunk will bring last 10 characters from previous chunk)
splitter = RecursiveCharacterTextSplitter(
    separators=[" ", "."],
    chunk_size=101,
    chunk_overlap=10,
)
```

```python
#split the text
splitter.split_text(selected_content)
```

```python
#split the text and transform into langchain Document object
splitter.create_documents([selected_content])
```
