```python
# load your api_key
from google.colab import userdata
openai_api_key = userdata.get('OPENAI_API_KEY')
```

```python
# try your first question
from openai import OpenAI
client = OpenAI(
    api_key=openai_api_key
)

response = client.responses.create(
    model="gpt-5.4-nano",
    input="Coba jelaskan secara singkat apa yang dimaksud dengan LLM"
)

print(response.output_text)
```

```python
# raw response
response
```

```python
# access your token usage
input_tokens = response.usage.input_tokens
output_tokens = response.usage.output_tokens

print(f"Input tokens: {input_tokens}")
print(f"Output tokens: {output_tokens}")
```

```python
# calculate the price
price_usd = (input_tokens*0.2 + output_tokens*1.25) / 1_000_000
price_idr = price_usd * 17_000

print(f"Price in USD: {price_usd}")
print(f"Price in IDR: {price_idr}")
```
