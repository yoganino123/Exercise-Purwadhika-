# Why we need text processing?

**Raw text is messy and inconsistent** - just like how you need to clean vegetables before cooking, we need to clean text before analysis.
Problems with raw text:

- "Happy", "HAPPY", and "happy" are treated as 3 different words by computers
- "running", "runs", "ran" are all about the same action but look different
- Words like "the", "and", "is" appear everywhere but don't add meaning
- Typos, extra spaces, and symbols create noise

**What happens without preprocessing:**

- A computer might think "Good" and "good" are completely different
- "I'm running" and "I am running" would be seen as totally different sentences
- Important patterns get hidden in the mess

**What preprocessing achieves:**

- Consistency: All text follows the same format
- Focus: Keeps only meaningful words
- Efficiency: Reduces data size and processing time
- Better results: Machine learning models work much better on clean, standardized text

**Real example:**

- Raw: "I'm REALLY loving this Amazing product!!! 😍"
- Processed: "really love amazing product"
- Now the computer can easily find patterns and compare with other similar sentences

## A. Character-level preprocessing

````python
# Replace a punctuation only
re.sub('!', '%', original_text)
````

````python
# remove email in text
text_sample = "Email saya adalah rusnandifikri@gmail.com"
re.sub(r'\S+@\S+', '', text_sample)
````

````python
# Remove URL
text_sample = "Website saya adalah https://www.google.com"
re.sub(r'http\S+', '', text_sample)
````

## C. Token Preprocessing

Token preprocessing involves cleaning and standardizing individual tokens (words) after tokenization. The three main techniques are:

- **Stop Words Removal**: Remove common words that don't carry meaning
- **Stemming**: Reduce words to their root form (crude method)
- **Lemmatization**: Reduce words to their dictionary form (smart method)

### 🛑 Stop Words Removal

#### What it does
Removes common words like "the", "and", "is", "are" that appear frequently but don't contribute to meaning.

#### When to use

##### ✅ Always use for:
- **Search engines**: Focus on meaningful search terms
- **Topic modeling**: Identify key themes without noise
- **Keyword extraction**: Find important words only

##### ❌ Skip for:
- **Grammar checking**: Grammar rules need all words
- **Language translation**: Word order and articles matter
- **Sentiment analysis (sometimes)**: Words like "not" are crucial

##### 💡 Why:
Stop words rarely carry meaning but take up space and processing time.

#### Example
```
Before: ["I", "am", "running", "through", "the", "beautiful", "garden"]
After:  ["running", "beautiful", "garden"]
```

````python
original_text = """Hello world! How are you today? I'm doing great.
Natural Language Processing is fascinating. Let's learn tokenization together!"""
````

````python
from nltk.corpus import stopwords
````

````python
stop_words = set(stopwords.words('english'))
stop_words
````

````python
tokens = word_tokenize(original_text)
# Remove stop words from our tokens
filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
````

````python
filtered_tokens
````

````python
print(f"Before:\n{original_text}")
````

````python
filtered_text = " ".join(filtered_tokens)
print(f"After:\n{filtered_text}")
````

### ✂️ Stemming

#### What it does
Cuts off word endings to get the root form using simple rules.

#### When to use

##### ✅ Good for:
- **Quick prototyping**: Fast development and testing
- **Search applications**: Find all forms of a word quickly
- **Large datasets**: When speed is more important than accuracy

##### ❌ Avoid for:
- **Final production systems**: May create non-words
- **When accuracy matters**: Can be too aggressive

##### 💡 Why:
Fast but can create meaningless word fragments like "beauti" from "beautiful".

#### Example
```
running → run
beautiful → beauti  (not a real word!)
studies → studi
```

````python
# Example words for stemming
original_tokens = ['running', 'runs', 'ran', 'runner', 'easily', 'fairly', 'quickly',
               'beautiful', 'beautifully', 'flowers', 'flowering', 'blooming', 'bloomed']
print(f"Before:\n{original_tokens}")
````

````python
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in original_tokens]
print(f"After:\n{stemmed_tokens}")
````

### 📚 Lemmatization

#### What it does
Uses dictionary and grammatical analysis to find the proper root form of words.

#### When to use

##### ✅ Good for:
- **Production systems**: When you need real, meaningful words
- **Academic research**: Requires linguistic accuracy
- **High-quality analysis**: Better results worth the extra time

##### ⚠️ Consider cost:
- **Takes more time**: Slower than stemming
- **Requires more memory**: More complex processing

##### 💡 Why:
Produces real words, more linguistically accurate than stemming.

#### Example
```
running → run
better → good
mice → mouse
children → child
```

````python
original_tokens = ['running', 'runs', 'ran', 'better', 'good', 'children', 'feet',
                'mice', 'geese', 'beautiful', 'quickly', 'flowers', 'blooming']
print(f"Before:\n{original_tokens}")
````

````python
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word.lower()) for word in original_tokens]
print(f"After:\n{lemmatized_tokens}")
````

### ⚠️ Rule Based Preprocessing

Due to the limitations of corpus data and linguistic knowledge in the libraries or models used, there may be incorrect lemmatization and stemming results. This is why we need to incorporate rule-based approaches or manual corrections with strong domain expertise.

````python
original_tokens = ['running', 'runs', 'ran', 'better', 'good', 'children', 'feet',
                'mice', 'geese', 'beautiful', 'quickly', 'flowers', 'blooming']
print(f"Before:\n{original_tokens}")
lemmatized_tokens = [lemmatizer.lemmatize(word.lower()) for word in original_tokens]
print(f"After:\n{lemmatized_tokens}")
word_corrections = {
    'running': 'run',
    'better': 'good',  # correct but domain-specific handling
    'ran': 'run',      # correct but might need context
}
word_corrections_tokens = [word_corrections[word] if word in word_corrections else word for word in lemmatized_tokens]
print(f"After correction:\n{word_corrections_tokens}")
````

## 🚀 Quick Decision Guide

### For beginners or quick projects:
1. **Always** remove stop words
2. Use **stemming** for speed
3. Upgrade to **lemmatization** when accuracy matters

### For production systems:
1. **Always** remove stop words (unless grammar/translation)
2. Use **lemmatization** for better quality
3. Consider the trade-off: accuracy vs speed

### For specific applications:

| Application | Stop Words | Method | Reason |
|-------------|------------|--------|---------|
| Search Engine | ✅ Remove | Stemming | Speed matters |
| Chatbot | ✅ Remove | Lemmatization | Accuracy matters |
| Topic Modeling | ✅ Remove | Lemmatization | Quality topics |
| Sentiment Analysis | ⚠️ Maybe | Lemmatization | "not happy" ≠ "happy" |
| Grammar Checker | ❌ Keep | Lemmatization | Need all words |

---

## 💡 Pro Tips

1. **Order matters**: Tokenize → Remove Stop Words → Stem/Lemmatize
2. **Test both methods**: See which works better for your specific use case
3. **Consider your data**: Informal text (social media) might need different preprocessing
4. **Monitor performance**: Lemmatization is slower but often worth it
5. **Custom stop words**: Add domain-specific words to your stop word list

---

## 🎯 Key Takeaway

> **Start simple, then optimize**: Begin with stop word removal + stemming, then upgrade to lemmatization when you need better accuracy. The "best" method depends on your specific use case and requirements.
