# Word Embeddings: Why They Matter More Than BoW and TF-IDF

## Overview

This README explains why word embeddings have become the preferred method for text representation in modern NLP, comparing them with traditional approaches like Bag of Words (BoW) and Term Frequency-Inverse Document Frequency (TF-IDF).

## What Are Word Embeddings?

Word embeddings are dense vector representations of words in a continuous vector space, where semantically similar words are positioned close to each other. Unlike traditional sparse representations, embeddings capture semantic relationships and contextual meaning.

**Key characteristics:**
- Dense vectors (typically 100-300 dimensions)
- Capture semantic similarity
- Learn contextual relationships
- Continuous vector space representation

## Traditional Methods: BoW and TF-IDF

### Bag of Words (BoW)
- Represents text as a collection of individual words
- Ignores word order and grammar
- Creates sparse, high-dimensional vectors
- Each dimension represents a unique word in the vocabulary

### TF-IDF (Term Frequency-Inverse Document Frequency)
- Extends BoW by weighting terms based on their importance
- TF: How frequently a term appears in a document
- IDF: How rare or common a term is across all documents
- Still creates sparse, high-dimensional representations

## Why Word Embeddings Are Superior

### 1. **Semantic Understanding**

**Traditional methods:**
```
"king" and "queen" = completely different vectors
"car" and "automobile" = no relationship captured
```

**Word embeddings:**
```
Similar words have similar vector representations
king - man + woman ≈ queen (famous analogy)
"car" and "automobile" vectors are close in space
```

### 2. **Dimensionality and Efficiency**

| Method | Typical Dimensions | Sparsity |
|--------|-------------------|-----------|
| BoW | 10,000 - 100,000+ | 99%+ sparse |
| TF-IDF | 10,000 - 100,000+ | 99%+ sparse |
| Word Embeddings | 100 - 300 | Dense (0% sparse) |

**Benefits:**
- Lower memory requirements
- Faster computation
- Better generalization

### 3. **Context Awareness**

**BoW/TF-IDF limitations:**
- "Apple the company" vs "apple the fruit" - same representation
- Word order completely ignored
- No understanding of context

**Word embeddings advantages:**
- Context-dependent representations (especially with modern transformers)
- Capture syntactic and semantic relationships
- Handle polysemy (multiple meanings) better

### 4. **Handling Unknown Words**

**BoW/TF-IDF:**
- Unknown words = zero vectors
- No way to infer meaning from similar words

**Word embeddings:**
- Can infer representations for unknown words
- Subword embeddings handle out-of-vocabulary terms
- Transfer learning from pre-trained models

### 5. **Mathematical Properties**

Word embeddings enable powerful mathematical operations:

```python
# Vector arithmetic examples
king - man + woman ≈ queen
Paris - France + Italy ≈ Rome
running - run + walk ≈ walking
```

These relationships are impossible with sparse BoW/TF-IDF representations.

## When to Use Each Method

### Use BoW/TF-IDF when:
- Simple document classification tasks
- Interpretability is crucial
- Limited computational resources
- Working with very domain-specific vocabulary
- Quick prototyping needed

### Use Word Embeddings when:
- Semantic understanding is important
- Working with synonyms and related terms
- Building modern NLP applications
- Need to capture word relationships
- Working with multiple languages
- Handling large vocabularies efficiently

# Word2Vec

**sg** (Skip-gram vs CBOW architecture)
 - 0: **CBOW** - Continuous Bag of Words predicts the target word from surrounding context words. Faster training, works well with frequent words.
 - 1: **Skip-gram** - Predicts surrounding context words from a target word. Better performance on infrequent words and larger datasets.

**hs** (Training algorithm)
 - 0: **Negative sampling** - Approximates softmax by sampling a few negative examples instead of computing probabilities for entire vocabulary. More efficient and often better quality embeddings.
 - 1: **Hierarchical softmax** - Uses a binary tree structure to reduce computational complexity from O(V) to O(log V). Better for frequent words, less memory intensive.

````python
model = Word2Vec(sentences=sentences, vector_size=128, window=5, min_count=3, workers=4, epochs=100)
````

````python
# Save Model

os.makedirs("model/w2v", exist_ok=True)
model.save("model/w2v/gojek_review.w2v")
````

````python
# Load Model

model = Word2Vec.load("model/w2v/gojek_review.w2v")
````

````python
# Continue Training

train_examples = [
    ["aplikasinya", "ok", "kok", ".", "sangat", "membantu", "untuk", "keseharian", "saya"],
    ["aplikasi", "gojek", "ini", "adalah", "aplikasi", "paling", "kacau", "yang", "pernah", "saya", "coba"]
]
````

````python
model.train(corpus_iterable=train_examples, total_examples=len(train_examples), epochs=1)
````

# FastText vs Word2Vec - Key Differences

**Subword Information**

- Word2Vec: Treats each word as an atomic unit. Unknown words = no representation.
- FastText: Breaks words into character n-grams (subwords). Can handle unknown words by combining subword vectors.

**Out-of-Vocabulary (OOV) Problem**

- Word2Vec: Cannot represent words not seen during training.
- FastText: Can generate embeddings for new/misspelled words using subword components.

**Morphology Awareness**

- Word2Vec: "running" and "run" are completely separate entities.
- FastText: Understands that "running" and "run" share subword patterns, creating related embeddings.

**Example:**

- Word not in training: "unhappiness"
- Word2Vec: No embedding available
- FastText: Combines subwords like "un-", "happy", "-ness" to create a meaningful representation

**Trade-off:** FastText is more robust but uses more memory and can be slower due to subword processing.
