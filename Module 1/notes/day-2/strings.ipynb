```python
# kutip dua

teks = 'He said, "Hello!"'
print(teks)
```

```python
# kutip satu

teks = "It's a beautiful day! Hello!"
print(teks)
```

```python
# kutip tiga

teks = '''It's a beautiful day!
He said, "Hello!"



Have a nice day!'''
print(teks)
```

```python
prompt = '''User: Hello, how are you?
Assistant: I'm fine, thank you! How can I assist you today?'''

print(prompt)
```

```python
# escape character

teks = "He said, \"It's a beautiful day!\""
print(teks)
```

```python
# len function

teks = "Python is a programming language that lets you work quickly and integrate systems more effectively."
panjang_teks = len(teks) # Menghitung panjang teks dalam karakter
print(panjang_teks)
```

```python
# lowercase
teks = "Hello, WORLD!"
teks_lower = teks.lower()
print(teks_lower)
```

```python
# uppercase
teks = "Hello, world!"
teks_upper = teks.upper()
print(teks_upper)
```

```python
# split
teks = "Python is fun"
kata_kata = teks.split()
kata_kata
```

```python
# split dengan pemisah khusus
teks = "Python-is fun"
kata_kata = teks.split("-")
kata_kata
```

# string format

```python
nama = "Andi"
umur = '25'
teks = "Halo nama saya " + nama + " dan saya berumur " + umur + " tahun."
print(teks)
```

```python
# string format

nama = "Fikri"
umur = 30
teks = f"Halo nama saya {nama} dan saya berumur {umur} tahun."
print(teks)
```

```python
nama_chatbot = "Purwadhika Chatbot"

prompt = f"""User: Halo, siapa namamu?
Assistant: Nama saya {nama_chatbot}. Senang bertemu denganmu!"""

print(prompt)
```

```python
# check string

username = "Rusnandi Fikri"

x = "Fikri" in username
print(x)
```

```python
# check string

username = "Rusnandi Fikri"

x = "Budi" in username
print(x)
```

```python
# usecase mencari nama user
users = ["Alex Budi",  "Rusnandi Fikri", "John Doe", "John Fikri"]

for name in users:
    if "Fikri" in name:
        print(f"Nama {name} ditemukan dalam daftar users.")
```
