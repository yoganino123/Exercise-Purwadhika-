# Comparison

```python
penghasilan > pengeluaran # Bernilai True
```

```python
penghasilan <= pengeluaran # Bernilai False
```

```python
umur1 = 25
umur2 = 30
```

```python
umur1 == umur2  # Bernilai False
```

```python
umur1 != umur2  # Bernilai True
```

# Logical Operator

```python
penghasilan = 500_000
pengeluaran = 300_000
```

```python
# logical AND
penghasilan > 400_000 and pengeluaran < 400_000  # True and True -> True
```

```python
penghasilan > 400_000 and pengeluaran < 100_000  # True and False -> False
```

```python
umur_budi = 25
umur_rudi = 30
```

```python
# logical OR
umur_budi < 20 or umur_rudi > 15 # False or True -> True
```

```python
umur_budi < 20 or umur_rudi < 20 # False or False -> False
```

```python
# operator not

pendidikan = "SMA"
```

```python
pendidikan != "S1"  # Bernilai True
```

```python
warna = ["merah", "biru", "hijau"]
"kuning" not in warna  # Bernilai True
```

```python
"merah" not in warna  # Bernilai False
```

# Conditional Statement IF

```python
jumlah_pembelian = 1

if jumlah_pembelian > 2:
    print("Anda mendapatkan diskon 10%")
```

```python
jumlah_pembelian = 1

if jumlah_pembelian > 2:
    print("Anda mendapatkan diskon 10%")
else:
    print("Jumlah pembelian kurang dari 2, tidak mendapatkan diskon")
```

```python
nilai = 77
if nilai >= 80:
    print("Grade A")
elif nilai >= 70:
    print("Grade B")
else:
    print("Grade C")
```

```python
nilai = 52
if nilai >= 80:
    print("Grade A")
elif nilai >= 70:
    print("Grade B")
elif nilai >= 60:
    print("Grade C")
elif nilai >= 50:
    print("Grade D")
else:
    print("Grade E")
```

```python
nilai_akhir = 80
if nilai_akhir >= 75:
    print("Selamat, Anda lulus!")
    print("Anda lulus dengan predikat sangat baik")
else:
    print("Maaf, Anda tidak lulus.")
```

```python
nilai_uas = 70
nilai_uts = 60
nilai_akhir = (nilai_uas + nilai_uts) / 2
if nilai_akhir >= 75:
    if nilai_uas >= 80:
        print("Lulus dengan predikat sangat baik")
    else:
        print("Lulus")
else:
    print("Tidak lulus")
```
