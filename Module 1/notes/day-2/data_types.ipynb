# Data Types Python

Berikut ini adalah catatan pembelajaran tentang data types python.

```python
# string
kota = "Jakarta"
type(kota)
```

```python
# numeric - int
umur = 25
type(umur)
```

```python
# numeric - float
berat = 70.5
type(berat)
```

```python
# collection - list
buah = ["apel", "pisang", "jeruk"]
type(buah)
```

```python
# collection - tuple
warna = ("merah", "hijau", "biru")
type(warna)
```

```python
# collection - dictionary
data_diri = {"nama": "Andi", "usia": 30, "kota": "Bandung"}
type(data_diri)
```

```python
# set
angka_set = {1, 2, 3, 4, 5}
type(angka_set)
```

```python
# boolean
is_student = True
type(is_student)
```

```python
# None
nilai = None # --> missing value
type(nilai)
```
