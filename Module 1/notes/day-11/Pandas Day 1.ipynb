# Series and Dataframe

```python
import pandas as pd
import numpy as np
```

```python
# Series from list
mydata = [0,1,2,3,4]
pd.Series(mydata)
```

```python
# Series from numpy array
mydata = np.array([0,2,3,4,5,7])
pd.Series(data=mydata)
```

```python
sample = {
    'a':10,
    'b':20,
    'c':30
}
pd.Series(sample)
```

```python
# Dataframe from list
sample = [[1,2,3], [4,5,6], [7,8,9]]
pd.DataFrame(sample)
```

```python
sample = [[1,2,3], [4,5,6], [7,8,9]]
pd.DataFrame(data=sample, columns=["tinggi_badan", "berat_badan", "usia"])
```

```python
# Dataframe from numpy array
sample = np.random.randn(5,4)
pd.DataFrame(sample)
```

```python
# Dataframe from dict
sampel = {
    'a': [1,2,3],
    'b': [4,5,6],
}
pd.DataFrame(sampel)
```

# Selection and Indexing

```python
sample = np.random.randn(5,4)
df = pd.DataFrame(sample, columns=["tinggi_badan", "berat badan", "usia", "penghasilan"])
df
```

```python
# selecting a column
df.tinggi_badan
```

```python
df["berat badan"]
```

```python
# Accessing data by index
df['tinggi_badan'][0]
```

```python
# Accessing range in column
df['tinggi_badan'][2:]
```

```python
# Accessing 2 or more columns
df[['tinggi_badan', 'usia']]
```

```python
df = pd.DataFrame(sample, columns=["tinggi_badan", "berat badan", "usia", "penghasilan"], index=['a','b','c','d','e'])
df
```

## conditional selection

```python
sample = np.random.randn(10,4)
df = pd.DataFrame(sample, columns=["W", "X", "Y", "Z"])
df
```

```python
# conditional selection by dataframe
df[df>0]
```

```python
# conditional selection by columns
df[df['W']>0]
```

```python
df[df['W']>0][['Y','Z']]
```

```python
# multiple condition selection
df[(df['W']>0) & (df['Y']>0)]
```

```python
employee = {
    'name': ['John Wick', 'Jane Stone', 'Jack Sparrow', 'Jill Smith'],
    'gender': ['M', 'F', 'M', 'F'],
    'hire_date': ['2020-01-01', '2020-02-01', '2020-03-01', '2020-04-01'],
    'salary': [500, 600, 700, 800]
}
df = pd.DataFrame(employee)
df
```

```python
df[df['name'] == 'John Wick']
```

# Manipulating Data

```python
df.drop('new', axis=1)
```

```python
df
```

```python
df1 = df.drop('new', axis=1)
df1
```

```python
df.drop('new', axis=1, inplace=True)
```
