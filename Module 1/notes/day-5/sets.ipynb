## Subset, Superset, and Disjoint

```python
# Cek apakah dua set tidak memiliki elemen yang sama
indonesia = {'Jakarta', 'Bandung', 'Surabaya'}
eropa = {'London', 'Berlin', 'Madrid'}

indonesia.isdisjoint(eropa)
```

```python
indonesia = {'Jakarta', 'Bandung', 'Surabaya', 'Pontianak', 'Palangkaraya'} #superset
pulau_jawa = {'Jakarta', 'Bandung', 'Surabaya'} #subset
```

```python
pulau_jawa.issubset(indonesia)
```

```python
indonesia.issubset(pulau_jawa)
```

```python
indonesia.issuperset(pulau_jawa)
```
