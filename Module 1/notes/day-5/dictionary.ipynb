## Looping in Dictionary

```python
data = {
    'nama': 'Budi',
    'umur': 30,
    'kota': 'Bandung',
    'pekerjaan': 'Dokter'
}
data
```

```python
data.keys()  # Mendapatkan semua kunci dalam dictionary
```

```python
for key in data.keys():
    print(key)
```

```python
data.values()  # Mendapatkan semua nilai dalam dictionary
```

```python
for value in data.values():
    print(value)
```

```python
data.items()  # Mendapatkan semua pasangan kunci-nilai dalam dictionary
```

```python
for item in data.items():
    print(item)
```

```python
for key,val in data.items():
    print(f'Key: {key}, Value: {val}')
```

```python
list_keys = list(data.keys())  # Mengonversi kunci dictionary menjadi list
list_keys
```

```python
'umur' in list_keys # Memeriksa apakah 'umur' ada dalam daftar kunci
```

```python
'umur' in data.keys()
```

```python
'Budi' in data.values()
```
