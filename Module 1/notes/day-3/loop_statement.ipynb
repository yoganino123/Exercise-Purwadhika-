# Break condition

```python
password_benar = 'purwadhika123'

while True:
    password = input("Masukkan password: ")
    if password == password_benar:
        print("Selamat datang di Purwadhika!")
        break
    else:
        print("Password salah, silakan coba lagi.")
```

```python
total = 0

for angka in range(1, 101):
    total += angka
    print(f"Menambahkan {angka}, total sekarang: {total}")
    
    if total > 50:
        print("Batas tercapai! Menghentikan perhitungan.")
        break
```

```python
daftar_tamu = ["Andi", "Budi", "Caca", "Deni", "Euis"]
target = "Caca"

for tamu in daftar_tamu:
    print(f"Memeriksa: {tamu}")
    
    if tamu == target:
        print(f"Ketemu! {target} ada di daftar.")
        break  # Loop berhenti di 'Caca', 'Deni' dan 'Euis' tidak akan diperiksa
```

# Continue Condition

```python
angka = 0

while angka < 10:
    angka += 1
    
    # # Jika angka habis dibagi 2 (genap), lewati baris print di bawah
    # if angka % 2 == 0:
    #     continue 
        
    print(f"Angka ganjil: {angka}")
```

```python
angka = 0

while angka < 10:
    angka += 1
    
    # Jika angka habis dibagi 2 (genap), lewati baris print di bawah
    if angka % 2 == 0:
        continue 
        
    print(f"Angka ganjil: {angka}")
```

```python
daftar_nilai = [85, 40, 92, 55, 78, 60]

print("Daftar Nilai yang Lulus:")

for nilai in daftar_nilai:
    if nilai < 70:
        continue  # Skip nilai ini, langsung ambil angka berikutnya di list
    
    print(f"- Selamat! Nilai {nilai} dinyatakan Lulus.")
```

```python
angka = [64, 34, 25, 12, 22, 11, 90]
n = len(angka)

# Loop pertama untuk mengontrol berapa kali pengecekan dilakukan
for i in range(n):
    # Loop kedua untuk membandingkan angka bersebelahan
    # (n-i-1) digunakan karena angka di ujung kanan sudah pasti urut
    for j in range(0, n - i - 1):
        
        # Cek jika angka kiri > angka kanan
        if angka[j] > angka[j + 1]:
            # Tukar posisi angka
            angka[j], angka[j + 1] = angka[j + 1], angka[j]

print("Hasil urut:", angka)
```
