# Loop Through a List

```python
barang = ["pensil", "buku", "penghapus", "spidol", "penggaris"]
```

```python
for item in barang:
    print("Saya membeli", item)
```

```python
print(len(barang))
```

```python
for i in range(len(barang)):
    if barang[i] == "spidol":
        print("Saya menemukan spidol di indeks ke-", i)
        del barang[i]
        break
```

```python
barang
```

```python
barang = ["pensil", "buku", "penghapus", "spidol", "penggaris"]
```

```python
"penghapus" in barang
```

```python
if "penghapus" in barang:
    print("Saya menemukan penghapus di dalam daftar barang.")
    barang.remove("penghapus")
    print("Penghapus telah dihapus dari daftar barang.")
else:
    print("Penghapus tidak ditemukan dalam daftar barang.")
```

# Find Index

```python
buah = ["apel", "pisang", "jeruk", "mangga", "semangka"]
```

```python
buah.index("jeruk") # Mengembalikan indeks dari elemen "jeruk"
```

```python
if "semangka" in buah:
    print("Indeks buah tersebut adalah:", buah.index("semangka"))
else: 
    print("Buah tersebut tidak ditemukan dalam daftar.")
```

```python
namabuah = "mangga"

if namabuah in buah:
    print("Indeks buah tersebut adalah:", buah.index(namabuah))
else: 
    print("Buah tersebut tidak ditemukan dalam daftar.")
```
