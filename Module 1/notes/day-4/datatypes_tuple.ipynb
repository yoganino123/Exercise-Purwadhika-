```python
myTuple = ("hello", 42, 3.14, False)
myTuple
```

```python
myTuple[0]  # Mengakses elemen pertama
```

```python
myTuple[0:3]  # Mengakses elemen dari indeks 0 hingga 2 (3 tidak termasuk)
```

```python
len(myTuple)  # Mendapatkan panjang tuple
```

# Editing Data Tuple

```python
myTuple[0] = "world"  # Ini akan menghasilkan error karena tuple tidak dapat diubah
```

```python
myList = list(myTuple) # Mengonversi tuple menjadi list
myList
```

```python
myList[0] = "world"  # Sekarang kita dapat mengubah elemen karena ini adalah list
```

```python
myTuple = tuple(myList)  # Mengonversi kembali list menjadi tuple
myTuple
```

# Loop in Tuple

```python
buah = ("apel", "jeruk", "mangga")

for item in buah:
    print(item)
```

# Check if Item in Tuple

```python
"apel" in buah  # Memeriksa apakah "apel" ada dalam list buah
```

```python
"semangka" in buah  # Memeriksa apakah "semangka" ada dalam list buah
```

# Concatenating and Find Index

```python
buah1 = ("apel", "jeruk")
buah2 = ("mangga", "pisang")
```

```python
semua_buah = buah1 + buah2  # Menggabungkan dua tuple
semua_buah
```

```python
semua_buah.index("mangga")  # Mendapatkan indeks dari elemen "mangga"
```

# Tuple inside Tuple

```python
things = (
    ("apel", "mangga", "pisang"), # sub-tuple pertama
    (1, 2, 3, 4, 5),              # sub-tuple kedua
    (True, False)                 # sub-tuple ketiga
)
```

```python
things[0]  # Mengakses sub-tuple pertama
```

```python
things[0][0]    # Mengakses elemen pertama dari sub-tuple pertama
```

```python
things[1][1:4] # Mengakses elemen dari indeks 1 hingga 3 pada sub-tuple kedua
```
