# Pandas Operation

Data source : https://www.kaggle.com/datasets/amar5693/screen-time-sleep-and-stress-analysis-dataset/data

```python
data_path = '/content/drive/MyDrive/JCAIEAH-003/Notes and Hands On/day-12/Smartphone_Usage_Productivity_Dataset_50000.csv'
```

```python
import pandas as pd

# read csv
df = pd.read_csv(data_path)
```

```python
df.head()
```

```python
# listing all columns
df.columns
```

```python
# df info
df.info()
```

```python
# create new columns about average monthly phone hour
df["Monthly_Phone_Hours"] = df["Daily_Phone_Hours"].apply(lambda x : x*30)
```

```python
df.head()
```

```python
# implementing apply function in df

def det_age_category(age):
  if age < 25:
    category = "Young Adult"
  elif age <= 40:
    category = "Adult"
  else:
    category = "Senior Adult"
  return category

df["Age_Category"] = df["Age"].apply(det_age_category)
```

```python
df.head()
```

```python
df[['Age', 'Age_Category']]
```

```python
# create new columns differencing phone hour with sleep hout
df["Phone_vs_Sleep_Hours"] = df["Daily_Phone_Hours"] - df["Sleep_Hours"]
```

```python
df.head()
```

```python
df[['Daily_Phone_Hours', 'Sleep_Hours', 'Phone_vs_Sleep_Hours']]
```

```python
# select some columns from df
df1 = df[['Age_Category', 'Gender', 'Occupation', 'Daily_Phone_Hours']]
df1.head()
```

```python
pivot_table_df1 = df1.pivot_table(index=['Age_Category', 'Occupation'], columns='Gender', aggfunc='median', values='Daily_Phone_Hours')
display(pivot_table_df1)
```

```python
pivot_table_df1.reset_index(inplace=True)
pivot_table_df1
```

```python
pivot_table_df1.to_excel('/content/drive/MyDrive/JCAIEAH-003/Notes and Hands On/day-12/pivot_table_df1.xlsx', index=False)
```
