# Pandas Merge, Join, and Concat

```python
import pandas as pd
import numpy as np
```

```python
left_data = {
    'key' : ['K0', 'K1', 'K2', 'K3'],
    'A' : ['A0', 'A1', 'A2', 'A3'],
    'B' : ['B0', 'B1', 'B2', 'B3']
}

right_data = {
    'key' : ['K0', 'K1', 'K3', 'K4'],
    'C' : ['C0', 'C1', 'C3', 'C4'],
    'D' : ['D0', 'D1', 'D3', 'D4']
}
```

```python
left = pd.DataFrame(left_data)
right = pd.DataFrame(right_data)
```

```python
left
```

```python
right
```

## Merge

```python
# inner
pd.merge(left, right, on='key')
```

```python
# outter
pd.merge(left, right, on='key', how='outer')
```

```python
# left
pd.merge(left, right, on='key', how='left')
```

```python
# right
pd.merge(left, right, on='key', how='right')
```

## Join

```python
right.set_index('key', inplace=True)
left.set_index('key', inplace=True)
```

```python
left
```

```python
right
```

```python
left.join(right, how='inner')
```

```python
left.join(right, how='outer')
```

```python
left.join(right, how='left')
```

```python
left.join(right, how='right')
```

## Concat

```python
data1 = {
    'Nama': ['A', 'B', 'C', 'D'],
    'Usia': [10, 11, 12, 13],
    'Berat': [30, 31, 32, 33],
    'Tinggi': [40, 41, 42, 43]
}

data2 = {
    'Nama': ['E', 'F', 'G', 'H'],
    'Usia': [14, 15, 16, 17],
    'Berat': [34, 35, 36, 37],
    'Tinggi': [44, 45, 46, 47]
}

data3 = {
    'Nama': ['I', 'J', 'K', 'L'],
    'Usia': [18, 19, 20, 21],
    'Berat': [38, 39, 40, 41],
    'Tinggi': [48, 49, 50, 51]
}
```

```python
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
```

```python
pd.concat([df1, df2, df3])
```

```python
pd.concat([df1, df2, df3], axis=1)
```
