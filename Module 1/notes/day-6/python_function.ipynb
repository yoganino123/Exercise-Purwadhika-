### Callback Functions

```python
def tambah(num1, num2):
    return num1+num2

def kurang(num1, num2):
    return num1-num2

def kali(num1, num2):
    return num1*num2

def hitung(fnOperator, angka1, angka2):
    hasil = fnOperator(angka1, angka2)
    return hasil
```

```python
print(tambah(1,2))
```

```python
print(kurang(5,2))
```

```python
hitung(tambah,2,3)
```

```python
hitung(kurang,4,1)
```

```python
hitung(kali,3,6)
```

### Calling Other Function

```python
def kurang(num1, num2):
    return num1-num2

def kali(num1, num2):
    return num1*num2

def harga_total(harga_satuan, jumlah, diskon):
    hargaTotal = kali(harga_satuan,jumlah)
    hargaAkhir = kurang(hargaTotal,diskon)
    return hargaAkhir
```

```python
harga_pensil = 1000
jumlah_beli = 12
diskon_total = 2500

harga_total(harga_pensil, jumlah_beli, diskon_total)
```

### Recursive Function

```python
def coutdown(counter):
    print(counter)
    counter -= 1

    if counter >= 0:
        coutdown(counter)
```

```python
coutdown(10)
```

### Filter Function

```python
# filter angka genap

def genap(angka):
    return angka%2 == 0

def ganjil(angka):
    return angka%2 == 1
```

```python
listAngka = [1,2,3,4,5,6,7,8,9]
```

```python
list(filter(genap, listAngka))
```

```python
tuple(filter(ganjil, listAngka))
```
